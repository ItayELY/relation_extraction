```python
from IPython.lib.security import passwd



def findRel(doc):
  treated = [d for d in doc if d.text == "treated"][0]
  diseases = []
  nsubj = [c for c in treated.children if c.dep_ in ["nsubj", "nsubjpass"]][0]
  #print(treated)
  treatments = []
  head = [c for c in treated.children if c.dep_ == "nmod"]
  for h in head:
    current = h
    try:
      #print(h)
      pred = [c for c in h.children if c.dep_ == "amod"][0]
    except:
      pass
    try:
      pred_txt = pred.text + " "
    except:
      pred_txt = ""
    
    treatments.append(pred_txt + h.text)

    kids = [c for c in h.children if c.dep_ == "conj"]
    while len(kids) != 0:
      current = kids[0]
      try:
        pred = [c for c in current.children if c.dep_ == "amod"][0]
      except:
        kids = [c for c in current.children if c.dep_ == "conj"]
        continue
      try:
        pred_txt = pred.text + " "
      except:
        pred_txt = ""
    
      treatments.append(pred_txt + current.text)
      kids = [c for c in current.children if c.dep_ == "conj"]
    try:
      for ent in doc.ents:
        if ent.label_ == "DISEASE":
          if nsubj.text in ent.text:
            diseases.append(ent.text)
        #print(ent.text, ent.label_)

        #print("disease: " + diseases[0], "treatment: " + t)
      return diseases[0], treatments
    except:
      pass
#print(treatments)
#print(diseases)
```

```python
import json

relations_part2 = []
with open('results_treated_by.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile)
    for row in spamreader:


     # print(row)
      try:
        doc = nlp4(row[1])
        d, ts = findRel(doc)
        for t in ts:
          instances = {"condition": d, "treatment": t, "evidence": row[1]}
          relations_part2.append(json.dumps(instances))

      except Exception as e:
        #print(e)
        pass
with open('results_treated_with.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile)
    for row in spamreader:

     # print(row)
      try:
        doc = nlp4(row[1])
        d, ts = findRel(doc)
        for t in ts:
          instances = {"condition": d, "treatment": t, "evidence": row[1]}
          relations_part2.append(json.dumps(instances))

      except Exception as e:
        #print(e)
        pass
```
